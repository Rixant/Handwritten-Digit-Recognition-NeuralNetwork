```python
#define sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
   

def softmax(z):
        # Numerically stable with large exponentials
        exps = np.exp(z - z.max())
        return exps / np.sum(exps, axis=0,keepdims=True)
```

```python
#initialize the required parameters with array of appropriate dimensions
def parameters_initialization(input_nodes, h1, h2, output_nodes):
    
    #weights are initialized in each layer with random values
    np.random.seed(0)
    W1=np.random.randn(h1, input_nodes)*0.01
    W2=np.random.randn(h2, h1)*0.01
    W3=np.random.randn(output_nodes, h2)*0.01
    
    b1=np.zeros((h1,1))
    b2=np.zeros((h2,1))
    b3=np.zeros((output_nodes,1))
    
    parameters={
        "W1":W1,
        "W2":W2,
        "W3":W3,
        "b1":b1,
        "b2":b2,
        "b3":b3
    }
    
    return parameters
    
```

```python
def forward_propagation(input,parameters):
    
    
    Z1=np.dot(parameters["W1"],input)+parameters["b1"]
    A1=sigmoid(Z1)
    Z2=np.dot(parameters["W2"],A1)+parameters["b2"]
    A2=sigmoid(Z2)
    Z3=np.dot(parameters["W3"],A2)+parameters["b3"]
    A3=sigmoid(Z3)
    
    activations={
        "A1": A1,
        "A2": A2,
        "A3": A3,
        "Z1": Z1,
        "Z2": Z2,
        "Z3": Z3
        
    }
    
    return activations
    
```

```python
#calculate cost function
def cost_function(output,A3):
    cost=cost=(-1/m)*np.sum((output*np.log(A3))+(1-output*np.log(1-A3)))
    
    return cost
```

```python
def back_propagation(input,output,learning_rate,parameters, activations):
    grads={}
    
    dZ3=(activations["A3"]-output)*activations["A3"]*(1-activations["A3"])
    dW3=np.dot(dZ3,activations["A2"].T)
    db3 = np.sum(dZ3, axis=1, keepdims=True)
    
    dZ2=np.dot(parameters["W3"].T,dZ3)*activations["A2"]*(1-activations["A2"])
    dW2=np.dot(dZ2,activations["A1"].T)
    db2 = np.sum(dZ2, axis=1, keepdims=True)
    
    dZ1=np.dot(parameters["W2"].T,dZ2)*activations["A1"]*(1-activations["A1"])
    dW1=np.dot(dZ1,input.T)
    db1 = np.sum(dZ1, axis=1, keepdims=True)
    

    
    
    
    
    
    
    
    
    grads={
        'dW1':dW1,
        'dW2':dW2,
        'dW3':dW3,
        'db1': db1,
        'db2': db2,
        'db3': db3
    }
    

    return grads
    
    
```

```python
def update_parameters(parameters,grads):
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]
    
    b1 = parameters["b1"]
    b2 = parameters["b2"]
    b3 = parameters["b3"]
        
    dW1 = grads["dW1"]
    dW2 = grads["dW2"]
    dW3 = grads["dW3"]
    
    db1 = grads["db1"]
    db2 = grads["db2"]
    db3 = grads["db3"]
    
   
    
    W1 = W1-learning_rate*dW1
    W2 =  W2-learning_rate*dW2
    W3 = W3-learning_rate*dW3
    
    b1 = b1-learning_rate*db1
    b2 = b2-learning_rate*db2
    b3 = b3-learning_rate*db3

    
    parameters = {"W1": W1,
                  "W2": W2,
                  "W3": W3,
                  "b1":b1,
                  "b2":b2,
                  "b3":b3
                }
    
    return parameters
```

```python
epochs=10
parameters=parameters_initialization(input_nodes, h1, h2, output_nodes)
for epoch in range(epochs):
    print("Epoch: "+ str(epoch+1)+"/"+str(epochs))   
    activations=forward_propagation(input,parameters)
#     cost=cost_function(output,activations["A3"])
#     print(cost)
    grads=back_propagation(input,output,learning_rate,parameters,activations)
    parameters=update_parameters(parameters,grads)
    print("A3")
    print(activations["A3"].T.shape)
    print(activations["A3"].T)
    
    
```
